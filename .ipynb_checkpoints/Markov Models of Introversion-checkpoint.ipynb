```julia
#It's a two-state system - you're either chilling alone or w/ people
type StateSpace
    private_node
    public_node
    p_pub2prv
    p_pub2pub
    p_prv2pub
    p_prv2prv
end

s1 = StateSpace(1,2,0.5,0.5,0.5,0.5); #Non-sticky space (people in this space like to move around)
s2 = StateSpace(1,2,0.1,0.9,0.1,0.9); #Sticky space (people in this space like to stay put)
```

```julia
#Okay, let's get real
type Person
    state
    social_factor
end

#Move function; probabilities get modified by a person's preference for private nodes
function move(person, state_space, start_state)
    move_roll = rand()
        
        if start_state==state_space.public_node
        p_pub2pub = state_space.p_pub2pub * person.social_factor
            
            if move_roll < p_pub2pub
                return state_space.public_node
            else
                return state_space.private_node
            end
        
        elseif start_state==state_space.private_node
        p_prv2pub = state_space.p_prv2pub * person.social_factor
            
            if move_roll < p_prv2pub
                return state_space.public_node
            else
                return state_space.private_node
            end
        end
    end
```

```julia
#Generate people
p1 = Person(1,.1); #Extreme introvert
p2 = Person(1, .5); #Introvert
p3 = Person(1, 1); #Non-introvert
```

```julia
#Run simulation
iters = Array(Int32,0)
results1 = Array(Int32,0)
results2 = Array(Int32,0)
results3 = Array(Int32,0)


for x=1:100
    p1.state = move(p1, s1, p1.state)
    p2.state = move(p2, s1, p2.state)
    p3.state = move(p3, s1, p3.state)
    push!(iters, x)
    push!(results1, p1.state)
    push!(results2, p2.state)
    push!(results3, p3.state)
end
```

```julia
plot(x=iters,y=results1, Geom.line) #Extreme introvert, non-sticky space
```

```julia
plot(x=iters,y=results2, Geom.line) #Introvert, non-sticky space
```

```julia
plot(x=iters,y=results3, Geom.line) #Non-introvert, non-sticky space
```
